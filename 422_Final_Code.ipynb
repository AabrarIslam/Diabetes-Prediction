```python
import csv
import matplotlib.pyplot as plt

def check_balance(sample_counts, threshold=0.1):
    class_counts_list = list(sample_counts.values())
    majority_class_count = max(class_counts_list)
    minority_class_count = min(class_counts_list)
    balance_ratio = minority_class_count / majority_class_count

    if balance_ratio <= (1 - threshold):
        return "Unbalanced"
    else:
        return "Balanced"

def load_csv(file_path):
    class_counts = {}
    with open(file_path, 'r') as csvfile:
        csv_reader = csv.reader(csvfile)
        next(csv_reader)  # Skip the header row
        for row in csv_reader:
            class_label = row[-1]  # Assuming the last column contains the class label
            if class_label not in class_counts:
                class_counts[class_label] = 0
            class_counts[class_label] += 1
    return class_counts

csv_file_path = '/content/drive/MyDrive/BRACU/CSE422/results.csv'  # Change this to your CSV file path

class_counts = load_csv(csv_file_path)
balance_status = check_balance(class_counts)

print("Class balance status:", balance_status)

# Plot the class distribution
plt.figure(figsize=(10, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.title('Class Distribution')
plt.xlabel('Class Labels')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
print("Training accuracy (%) =", lr.score(x_train,y_train) * 100)
print("Testing accuracy (%) =", lr.score(x_test,y_test)*100)
```

```python
res.isnull().sum() # Checking null values (found)
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(res[['Age']])
res['Age'] = impute.transform(res[['Age']])
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(res[['BMI']])
res['BMI'] = impute.transform(res[['BMI']])
```

```python
res.isnull().sum() # Checking null values (not found any)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(res[["Age", "BMI", "Diabetes"]], c=res["Diabetes"] , figsize=(12,12), alpha=1)
```

```python
res = res.drop([cols[1], cols[5], cols[7], cols[8], cols[9], cols[10], cols[12]], axis=1)
res
```

```python
x = res.drop(['Diabetes'], axis=1)
y = res['Diabetes']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
print("Training accuracy (%) =", lr.score(x_train,y_train) * 100)
print("Testing accuracy (%) =", lr.score(x_test,y_test)*100)
```

```python
prec_score = Precision()
y_pred=lr.predict(x_test)
prec_score.update_state(y_pred, y_test)
p = prec_score.result().numpy()
print('Precision Score = %.3f' % p)

recall_score = Recall()
recall_score.update_state(y_pred, y_test)
r = recall_score.result().numpy()
print('Recall Score = %.3f' % r)
```

```python
y_pred_lr = lr.predict(x_test)
cutoff = 0.5
y_pred_classes = np.zeros_like(y_pred_lr)    # initialise a matrix full with zeros
y_pred_classes[y_pred > cutoff] = 1

y_test_classes = np.zeros_like(y_pred_lr)
y_test_classes[y_test > cutoff] = 1
LABELS = ['1', '0']

conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)

plt.figure(figsize =(12, 12))

sb.heatmap(conf_matrix, xticklabels = LABELS, yticklabels = LABELS, annot = True, fmt ="d");


plt.ylabel('True Class')

plt.xlabel('Predicted Class')
plt.rcParams.update({'font.size': 30})
plt.show()
```
